```python
import matplotlib.pyplot as plt
from skimage.segmentation import slic, mark_boundaries
from skimage import io

# Load an example image (replace 'example.jpg' with your image file)
CATEGORY = 'bottle'
img_path = f"/home/zhaoxiang/dataset/mvtec_anomaly_detection/{CATEGORY}/train/good/024.png"
image = io.imread(img_path)

# Define the number of desired superpixels
num_superpixels = 50

# Apply superpixel segmentation using SLIC
segments = slic(image, n_segments=num_superpixels, compactness=10)

# Visualize the result
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(image)
ax[0].set_title('Original Image')

# Mark superpixel boundaries on the original image
ax[1].imshow(mark_boundaries(image, segments))
ax[1].set_title('Image with Superpixel Boundaries')

plt.show()
```
